```python
 import numpy as np
import pandas as pd
file_path=pd.read_csv('richest.csv',delimiter=";")
df=pd.DataFrame(file_path)
df
```

```python
columns_to_clean=['Total Net Worth','$ Last Change'	,'$ YTD Change'	]
for x in columns_to_clean:
    df[x] = df[x].str.replace('$', '')
df
```

```python

def vv(x):
    if isinstance(x, str):
        if "K" in x:
            if len(x) > 1:
                return float(x.replace("K","")) * 0.000001
        elif "M" in x:
            if len(x) > 1:
                return float(x.replace("M","")) * 1000
        elif "B" in x:
            if len(x) > 1:
                return float(x.replace("B","")) * 100000
df['$ Last Change'] = df['$ Last Change'].apply(vv)

print(df)
```
